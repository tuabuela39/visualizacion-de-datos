```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')

ruta = "/content/drive/My Drive/ventas_composicion.csv"

df = pd.read_csv(ruta)

df.head()
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    data=df,
    x='Fecha',        # eje X: tiempo
    y='Porcentaje',   # eje Y: cuota de mercado
    hue='Categoría',  # una línea por cada categoría
    marker='o'
)

plt.title('Evolución de la cuota de mercado por categoría')
plt.xlabel('Fecha')
plt.ylabel('Porcentaje')
plt.xticks(rotation=45)  # rota fechas si están amontonadas
plt.legend(title='Categoría')
plt.tight_layout()
plt.show()
```
