```python
import plotly_express as px
```

```python
gapminder =  px.data.gapminder()
gapminder.head()
```

```python
gapminder2007 = gapminder.query('year == 2007')
gapminder2007.head()
```

```python
labels_dict = {"lifeExp":"Life Expectancy", "pop":"Population", "gdpPercap":"GDP per Capita"}
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", labels=labels_dict, )
fig.show()
```

```python
labels_dict = {"lifeExp":"Life Expectancy", "pop":"Population", "gdpPercap":"GDP per Capita (USD)"}
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", labels=labels_dict, color = "continent" )
fig.show()
```

```python
labels_dict = {"lifeExp":"Life Expectancy", "pop":"Population", "gdpPercap":"GDP per Capita (USD)"}
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", labels=labels_dict, color = "continent", size="pop", size_max=60, hover_name="country", log_x=True, facet_col="continent" )
fig.show()
```

```python
fig2 = px.choropleth(gapminder, locations= "iso_alpha", color="lifeExp", hover_name="country", animation_frame="year", projection="natural earth", labels=labels_dict)
fig2.show()
```
